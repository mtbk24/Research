__In this notebook I take take old plots that I've created and read them back into Python using the mpimg module provided within matplotlib.image. To learn more about how to do this, see a basic tutorial at: https://matplotlib.org/users/image_tutorial.html __

__I then read in multiple old plots and place them into a single plot using matplotlib's subplot feature. This can also be done using matplotlib's gridspec module, but we did not use it in this notebook. __

```python
"""
Created on Wed Aug 16 13:00:54 2017

@author: KimiZ
"""

from __future__ import division
from collections import OrderedDict
import os, glob

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
#import itertools

%matplotlib inline
```

```python
def get_burst_data(burst):
    '''
    get_burst_data(burst)
    
    PARAMETERS:
    ----------
    burst: str, Fermi trigger ID. I often just call this 'burst' or 'burstname'
                instead of trigger id. 
    
    RETURNS:
    ----------
    Returns the GRB data that may be needed for a specific burst. 
    Data returned in a tuple format:
    (redshift, start_duration, end_duration, fermi_detectors)
    
    These data types for each element returned are:
    redshift - float
    start_duration - float
    end_duration - float
    fermi_detectors - list of strings.
    
    EXAMPLES:
    ----------
    In [40]: burst = 'bn080916009' 
             get_burst_data(burst)
    Out[40]: (4.35, 1.28, 64.257, ['n3', 'n4'])
    
    
    '''
    filename      = '/Users/KimiZ/GRBs2/Sample/LATbursts.txt'
    # DATA TYPES OF THE BURST DATA, IN DICT FORMAT.
    dtypes = {'names':('trigname', 'redshift', 't90start', 't90stop', 'detectors'), 
                'formats':('S11','f4','f4','f4','S11')} 
    # READ IN DATA USING NUMPYS LOADTXT
    data  = np.loadtxt(filename, dtype=dtypes, delimiter='\t\t', 
                       skiprows=1, usecols=(2,3,4,5,6))
    data  = pd.DataFrame(data)
    # REDSHIFT, DURATION START, DURATION END, FERMI DETECTORS 
    z, t90start, t90stop, detectors = list( data[data['trigname'] == burst].values[0] )[1:]
    detectors = list(detectors.split(','))
    z = round(z, 3) 
    t90start = round(t90start, 4) 
    t90stop = round(t90stop, 4) 
    return z, t90start, t90stop, detectors


def load_image(burst):
    '''
    load_image(burst)
    
    PARAMETERS:
    ----------
    burst: str, Fermi trigger ID. I often just call this 'burst' or 'burstname'
                instead of trigger id. 
    
    RETURNS:
    ----------
    Returns an image of this GRB's light curve that was previously plotted.
    We want to reuse these images in future plots. 
    
    '''
    direc = '/Users/KimiZ/Python/Github/%s/images/'%burst
    os.chdir(direc)  # change directory 
    # find file(s) with lightcurve_NaI in its name. Use first file in the list.  
    Files = glob.glob('*lightcurve_NaI*')[0] 
    # read in file and generate its image in Python.
    Img = mpimg.imread(Files)  # read in file and generate its image in Python.
    os.chdir(basedir) # return to original directory.
    return Img 
```

*************************************************************************************

__Retrieve a GRB's Light Curve that was previously plotted.__

__Load it into matplotlib as an image.__

__Re-plot it and place text over it.__

*************************************************************************************



```python
basedir = '/Users/KimiZ/Python/Github/Research/Plotting/'
```

```python
plt.clf()  # clear all previous plots, if any. 

burst = 'bn090510016'
img = load_image(burst)      # load image 

dat = get_burst_data(burst)  # get data on this burst.

# img dimensions format for trimming:
#    img[top:bottom, left:right, ]
# the third dimension I left blank deals with image color.
fig = plt.imshow(X=img[161:1670, 120:2530, ])

# the image is read in as a plot, so if we don't want matplotlib to show x and y
# axis numbers, use the following to turn the axes numbers, ticks, and labels off.
fig.axes.get_xaxis().set_visible(False)
fig.axes.get_yaxis().set_visible(False)
plt.box(on=False)

plt.figtext(0.65, 0.65, 
            'z = %.3f\ntstart = %.3f s\nduration = %.3f s'%(dat[0], dat[1], dat[2]-dat[1]),  
            fontsize=12)
plt.figtext(0.15, 0.8, burst, fontsize=16)
plt.figtext(0.15, 0.7, '17', fontsize=16)
plt.tight_layout()
plt.show()    

```

This image appears fuzzy here, but comes out clear when it is saved. We will see this in a moment after we make subplots of these light curves. 

```python
# Gamma-ray bursts shown with Fermi trigger ID and their number in my dataset. 
# these Fermi trigger IDs and their corresponding numbers will be shown on the plots. 

bursts =    OrderedDict([(0, ['bn080916009', 5]), 
                         (1, ['bn090323002', 13]), 
                         (2, ['bn090328401', 14]), 
                         (3, ['bn090902462', 20]), 
                         (4, ['bn090926181', 21]), 
                         (5, ['bn091003191', 24]), 
                         (6, ['bn091208410', 28]), 
                         (7, ['bn100728095', 33]), 
                         (8, ['bn110731465', 42]), 
                         (9, ['bn130518580', 58]), 
                         (10, ['bn131108862', 65]), 
                         (11, ['bn131231198', 66])])
```




*************************************************************************************

__Take multiple GRB's NaI Light Curves and place them on a single plot using matplotlib's subplots feature.__


*************************************************************************************





```python
plt.clf()  # clear all plots, if any.




fig, ax = plt.subplots(4, 3, figsize=(9.4, 7.8))

# BECAUSE WE ALREADY HAVE TO FOR LOOPS, WE DO NOT LOOP THROUGH 
#   BURST IDS AND NUMBERS, BUT INSTEAD WE USE AN ITERATION OF
#   n TO RETRIEVE BURST IDS AND NUMBERS. 

n = 0
for j in range(0,3):      # 3 columns of subplots
    for i in range(0,4):  # 4 rows of subplots
        burst,num = bursts[n]
        
        img = load_image(burst)  # load image 
        # place GRB lightcurve image in corresponding subplot and trim it.
        # dimensions to start image: img[]
        f = ax[i,j].imshow(img[160:1670, 120:2530, ]) 
        
        # remove x and y axes from plots. 
        # images are read in as plots, which have axes labels, etc.
        f.axes.get_xaxis().set_visible(False)
        f.axes.get_yaxis().set_visible(False)
        f.axes.set_frame_on(False) # remove frame. 
        
        # write burst ID and its number on each corresponding subplot. 
        # typically we'd use plt.figtext, but its not available for these types
        #   of subplots. Instead, use what would be the same as plt.text and 
        #   use the transform attribute. 
        f.axes.text(0.125, 0.8, '%s\n%2i'%(burst, num), 
                    fontsize=13, 
                    transform = ax[i,j].transAxes,  # need this when using .text
                    horizontalalignment='left')

        dat = get_burst_data(burst)  # get data on this burst.
        
        f.axes.text(0.6, 0.75, 
            'z = %.3f\ntstart = %.1f s\nduration = %.1f s'%(dat[0], dat[1], dat[2]-dat[1]),  
            fontsize=8,
            transform = ax[i,j].transAxes,
            horizontalalignment='left')

        n=n+1
        
        
plt.tight_layout(pad=0.01, w_pad=0.01, h_pad=0.01)
outfile = os.path.join(basedir,'figures/GRB_lightcurve_subplots.pdf')
fig.savefig(outfile, dpi=250)
#os.system('open %s'%outfile)  # open newly created file. 
#os.system('rm %s'%outfile)  # remove newly created file. 

#plt.show()
```

```python
def load_image_2(burst, detector, model):
    detdir = ('GBMwLAT' if 'L' in detector else 'GBM')
    filename = '/Users/KimiZ/GRBs2/analysis/LAT/' + \
    '%s/PYXSPEC/%s/%s/Plots_%s_-01-_%s_-11.png'%(burst, detdir, model, model, detector)
    Img = mpimg.imread(filename)
    return Img
```

```python
plt.clf()

burst = 'bn090510016'
img = load_image_2(burst, 'L', 'grbm')    
fig = plt.imshow(img)

# turn of image plot's axes. This is labels, ticks, etc.
fig.axes.get_xaxis().set_visible(False)
fig.axes.get_yaxis().set_visible(False)

plt.figtext(0.65, 0.85, burst, fontsize=16, 
            horizontalalignment='left')
plt.figtext(0.65, 0.8, '4', fontsize=16,
            horizontalalignment='left')

#plt.axes.set_frame_on(False)
plt.box(on=False)
plt.tight_layout()
plt.show()    
```

```python
plt.clf()


fig, ax = plt.subplots(4, 3, figsize=(9, 7.56))

n = 0
for j in range(0,3):
    for i in range(0,4):
        
        burst,num = bursts[n]
        
        img = load_image_2(burst, 'L', 'grbm')
        f = ax[i,j].imshow(img[75:1535, 90:2380, ])
        f.axes.get_xaxis().set_visible(False)
        f.axes.get_yaxis().set_visible(False)
        
        f.axes.set_frame_on(False)
        
        f.axes.text(0.9, 0.8, '%s\n%2i'%(burst, num), 
                    fontsize=10, 
                    transform = ax[i,j].transAxes,  # need this when using .text
                    horizontalalignment='right')
        n=n+1

plt.tight_layout(pad=0.05, w_pad=0.05, h_pad=0.05) 

outfile = os.path.join(basedir, 'figures/Band_XSPEC_nufnu_subplots.pdf')
fig.savefig(outfile, dpi=450)
os.system('open %s'%outfile)  # open newly created file. 
# os.system('rm %s'%outfile)  # remove newly created file. 

#plt.show()
```
