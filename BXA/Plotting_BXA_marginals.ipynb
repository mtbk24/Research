```python
def get_ParNames(model):
    '''
    Given a model name, it will return the parameter names of the model.
    See file 'ParameterNames.json in the $PYX/Analysis/ directory.
    '
    '''
    fname   = "/Users/KimiZ/GRBs2/analysis/pyxspec_programs/Analysis/ParameterNames.json"
    f       = json.load(open(fname, 'r'), encoding='utf-8')
    names   = f[model] # DICTIONARY OF PARAMETERS
    names   = [str(i) for i in names]
    return names

def marginal_plots_1donly(analyzer, XLims='5sigma', saveFigure=False):
    """
    Create marginal plots

    * analyzer: A instance of pymultinest.Analyzer
    * d: if more than 20 dimensions, by default only 1d marginal distributions
       are plotted. set d=2 if you want to force a 2d matrix plot

    """
    prefix      = analyzer.outputfiles_basename
    n_params    = analyzer.n_params
    parameters  = json.load(file(prefix + 'params.json'))
    a           = analyzer
    s           = a.get_stats()
    p           = PlotMarginal(a)  # moved a copy of file to this direc.

    #p = pymultinest.PlotMarginal(a) # call to /Users/KimiZ/BXA-master/bxa/xspec/plot.py

    values     = a.get_equal_weighted_posterior()
    assert n_params == len(s['marginals'])
    modes      = s['modes']
    
    plt.figure(figsize=(5*n_params, 5*n_params))
    from matplotlib.backends.backend_pdf import PdfPages
    print '1dimensional only. Set the D environment variable D=2 to force'
    print '2d marginal plots.'
    if saveFigure:
        pp = PdfPages(prefix + 'marg1d_KZ_wPYX.pdf')

    for i in range(n_params):
        plt.figure(figsize=(8, 8))
        plt.xlabel(parameters[i])

        m = s['marginals'][i]
        plt.xlim(m[XLims])

        oldax = plt.gca()
        x,w,patches = oldax.hist(values[:,i], bins=20, 
                                 edgecolor='grey', color='grey', 
                                 histtype='stepfilled', alpha=0.2)
        oldax.set_ylim(0, x.max())

        newax = plt.gcf().add_axes(oldax.get_position(), sharex=oldax, frameon=False)
        p.plot_marginal(i, ls='-', color='blue', linewidth=2)
        newax.set_ylim(0, 1)

        ylim = newax.get_ylim()
        y = ylim[0] + 0.05*(ylim[1] - ylim[0])
        center = m['median']
        low1, high1 = m['1sigma']
        print center, low1, high1
        newax.errorbar(x=center, y=y,
            xerr=numpy.transpose([[center - low1, high1 - center]]), 
            color='blue', linewidth=2, marker='s')

        bestFit = s['modes'][0]['maximum'][i]
        newax.vlines(bestFit, 0, 1, linestyle='--', color='black', 
                     lw=1, label="%.3f (BXA best)"%bestFit)

        PYXbestFit = PYX_bestfit["%s__%i"%(parameters[i], i+1)]
        newax.vlines(PYXbestFit, 0, 1, linestyle='--', color='brown', 
                     lw=1, label="%.3f (PYX best)"%PYXbestFit)

        newax.vlines(center, 0, 1, linestyle='-.', color='blue', 
                     lw=1, label="%.3f (BXA median)"%center)

        newax.hlines([0.15865, 1-0.15865], m[XLims][0],m[XLims][1], 
                     linestyle='-', color='dodgerblue', lw=1,)
        newax.vlines([low1, high1], ylim[0], ylim[1], 
                     linestyle='-', color='dodgerblue', lw=1,)

        oldax.set_yticks([])
        newax.set_ylabel("Probability")
        ylim = oldax.get_ylim()
        newax.set_xlim(m[XLims])
        oldax.set_xlim(m[XLims])
        plt.legend(loc='best', fontsize=8)
        if saveFigure:
            print('\n Saving Figure ... \n')
            plt.savefig(pp, format='pdf', bbox_inches='tight')
        else:
            plt.show()
    if saveFigure:
        pp.close()
 

def marginal_plots(analyzer, XLims='5sigma', saveFigure=False, d=None):
    """
    Create marginal plots

    * analyzer: A instance of pymultinest.Analyzer
    * d: if more than 20 dimensions, by default only 1d marginal distributions
       are plotted. set d=2 if you want to force a 2d matrix plot

    """
    prefix      = analyzer.outputfiles_basename
    n_params    = analyzer.n_params
    parameters  = json.load(file(prefix + 'params.json'))
    a           = analyzer
    s           = a.get_stats()
    p           = PlotMarginal(a)  # moved a copy of file to this direc.
    #p = pymultinest.PlotMarginal(a) # call to /Users/KimiZ/BXA-master/bxa/xspec/plot.py

    values     = a.get_equal_weighted_posterior()
    assert n_params == len(s['marginals'])
    modes      = s['modes']
    
    # IF YOU DON'T WANT THE BOTTOM HALF OF PLOTS, USE d = anything but 2
    # IF YOU WANT THE 2D MARGINALS, USE d=2 or leave it out.
    dim2 = ((1 if n_params > 20 else 2) if d is None else d) == 2

    if dim2:
        plt.figure(figsize=(5*n_params, 5*n_params))
        for i in range(n_params):
            plt.subplot(n_params, n_params, i + 1)
            plt.xlabel(parameters[i])

            m = s['marginals'][i]
            plt.xlim(m[XLims]) # used to be m['5sigma'], but those limits were too large. 

            oldax = plt.gca()
            x,w,patches = oldax.hist(values[:,i], bins=20, 
                                     edgecolor='grey', color='grey', 
                                     histtype='stepfilled', alpha=0.2)
            oldax.set_ylim(0, x.max())

            newax = plt.gcf().add_axes(oldax.get_position(), sharex=oldax, frameon=False)
            # MONOTONICALLY INCREASING BLUE PROBABILITY LINE.
            p.plot_marginal(i, ls='-', color='blue', linewidth=3)
            newax.set_ylim(0, 1)

            ylim = newax.get_ylim()
            y = ylim[0] + 0.05*(ylim[1] - ylim[0])
            # center, low1, high1 are the locations of the error bars and median.
            center = m['median']
            low1, high1 = m['1sigma']
            # ERRORBAR PLOT ON PROB. DISTRIBUTIONS. 
            newax.errorbar(x=center, y=y,
                xerr=numpy.transpose([[center - low1, high1 - center]]), 
                color='blue', linewidth=2, marker='s')
            for mode in modes:
                newax.errorbar(x = mode['mean'][i], 
                               y = 0.15, 
                               xerr = mode['sigma'][i], 
                               color='darkgoldenrod', lw=2, marker='s') 
            
            
            bestFit = s['modes'][0]['maximum'][i]
            
            newax.vlines(bestFit, 0, 1, linestyle='--', color='black', lw=1, 
                         label="%.3f (BXA best)"%bestFit)            
            
            PYXbestFit = PYX_bestfit["%s__%i"%(parameters[i], i+1)]
            newax.vlines(PYXbestFit, 0, 1, linestyle='--', color='brown', lw=1, 
                         label="%.3f (PYX best)"%PYXbestFit)

            newax.vlines(center, 0, 1, linestyle='-.', color='blue', lw=1, 
                         label="%.3f (BXA median)"%center)

            newax.hlines([0.15865, 1-0.15865], m[XLims][0],m[XLims][1], 
                         linestyle='-', color='dodgerblue', lw=1,)
            
            newax.vlines([low1, high1], ylim[0], ylim[1], 
                         linestyle='-', color='dodgerblue', lw=1,)

            oldax.set_yticks([])
            #newax.set_yticks([])
            newax.set_ylabel("Probability")
            ylim = oldax.get_ylim()
            newax.set_xlim(m[XLims])
            oldax.set_xlim(m[XLims])
            newax.legend(loc='best', fontsize=12)

            for j in range(i):
                plt.subplot(n_params, n_params, n_params * (j + 1) + i + 1)
                p.plot_conditional(i, j, bins=20, cmap = plt.cm.gray_r)
                # ERRORBARS ON LOWER PLOTS
                for mode in modes:
                    plt.errorbar(x=mode['mean'][i], 
                                 y=mode['mean'][j], 
                                 xerr=mode['sigma'][i], 
                                 yerr=mode['sigma'][j], 
                                 color='darkgoldenrod', lw=3, 
                                 label='x: %.3f $\pm$ %.3f \ny: %.3f $\pm$ %.3f'%(mode['mean'][i], mode['sigma'][i], mode['mean'][j], mode['sigma'][j]))
                plt.xlabel(parameters[i])
                plt.ylabel(parameters[j])
                plt.xlim(s['marginals'][i][XLims]) #['5sigma'])
                plt.ylim(s['marginals'][j][XLims]) #['5sigma'])
                plt.legend(loc='best', numpoints=None, fontsize=12)
                #plt.savefig('cond_%s_%s.pdf' % (params[i], params[j]), bbox_tight=True)
                #plt.close()
        plt.tight_layout() # kz keep axes from overlapping
        if saveFigure:
            print('\n Saving Figure ... \n')
            plt.savefig(prefix + 'marg_2.pdf')
        else:
            plt.show()
        

    else:
        pass
        from matplotlib.backends.backend_pdf import PdfPages
        print '1dimensional only. Set the D environment variable D=2 to force'
        print '2d marginal plots.'
        if saveFigure:
            pp = PdfPages(prefix + 'marg1d_KZ.pdf')

        for i in range(n_params):
            plt.figure(figsize=(5, 5))
            plt.xlabel(parameters[i])

            m = s['marginals'][i]
            #plt.xlim(m['5sigma'])
            plt.xlim(m[XLims])

            oldax = plt.gca()
            x,w,patches = oldax.hist(values[:,i], bins=20, edgecolor='grey', 
                                     color='grey', histtype='stepfilled', alpha=0.2)
            oldax.set_ylim(0, x.max())

            newax = plt.gcf().add_axes(oldax.get_position(), sharex=oldax, frameon=False)
            p.plot_marginal(i, ls='-', color='blue', linewidth=3)
            newax.set_ylim(0, 1)

            ylim = newax.get_ylim()
            y = ylim[0] + 0.05*(ylim[1] - ylim[0])
            center = m['median']
            low1, high1 = m['1sigma']
            print center, low1, high1
            newax.errorbar(x=center, y=y,
                xerr=numpy.transpose([[center - low1, high1 - center]]), 
                color='blue', linewidth=2, marker='s')
            
            for mode in modes:
                    newax.errorbar(x = mode['mean'][i], 
                                   y = 0.15, 
                                   xerr = mode['sigma'][i], 
                                   color='darkgoldenrod', lw=2, marker='s') 
            
            
            bestFit = s['modes'][0]['maximum'][i]
            newax.vlines(bestFit, 0, 1, linestyle='--', color='black', lw=1, 
                         label="%.3f (BXA best)"%bestFit)            
            PYXbestFit = PYX_bestfit["%s__%i"%(parameters[i], i+1)]
            newax.vlines(PYXbestFit, 0, 1, linestyle='--', color='brown', lw=1, 
                         label="%.3f (PYX best)"%PYXbestFit)

            newax.vlines(center, 0, 1, linestyle='-.', color='blue', lw=1, 
                         label="%.3f (BXA median)"%center)

            newax.hlines([0.15865, 1-0.15865], m[XLims][0],m[XLims][1], linestyle='-', 
                         color='dodgerblue', lw=1,)
            newax.vlines([low1, high1], ylim[0], ylim[1], linestyle='-', 
                         color='dodgerblue', lw=1,)
            

            oldax.set_yticks([])
            newax.set_ylabel("Probability")
            ylim = oldax.get_ylim()
            newax.set_xlim(m[XLims])
            oldax.set_xlim(m[XLims])
            plt.legend(loc='best')
            if saveFigure:
                print('\n Saving Figure ... \n')
                plt.savefig(pp, format='pdf', bbox_inches='tight')
            else:
                plt.show()
        if saveFigure:
            pp.close()
            
```
