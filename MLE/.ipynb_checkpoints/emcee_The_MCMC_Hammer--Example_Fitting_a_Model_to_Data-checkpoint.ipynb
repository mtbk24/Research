```python
from __future__ import division
from scipy.optimize import minimize, curve_fit
from math import pi
import numpy as np
from numpy import std, exp, log, log10


import matplotlib.pyplot as plt
%matplotlib inline
```

```python
line = lambda m,x,b: m*x+b
```

```python
# Choose the "true" parameters.
m_true = -0.9594
b_true = 4.294
f_true = 0.534

# Generate some synthetic data from the model.
N = 50
x = np.sort(10*np.random.rand(N))
yerr = 0.1+0.5*np.random.rand(N)
y = m_true*x+b_true
y += np.abs(f_true*y) * np.random.randn(N)
y += yerr * np.random.randn(N)
```

```python
# plot data
plt.figure(figsize=(9,8))
plt.errorbar(x = x, y = y,
             xerr = None,
             yerr = yerr,
             fmt = 'k.')
# plot model
plt.plot(x, line(m_true, x, b_true), ls='-', color='grey', label='true model')
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.legend(loc='best', numpoints=1)
plt.show()
```

# Least Squares Method

The true model is shown as the thick grey line and the effect of the underestimated uncertainties is obvious when you look at this figure. The standard way to fit a line to these data (assuming independent Gaussian error bars) is linear least squares. Linear least squares is appealing because solving for the parameters—and their associated uncertainties—is simply a linear algebraic operation. Following the notation in [Hogg, Bovy & Lang (2010)](https://arxiv.org/abs/1008.4686), the linear least squares solution to these data is

```python
A     = np.vstack((np.ones_like(x), x)).T
C     = np.diag(yerr * yerr)
cov   = np.linalg.inv(np.dot(A.T, np.linalg.solve(C, A)))

b_ls, m_ls = np.dot(cov, np.dot(A.T, np.linalg.solve(C, y)))
```

```python
# plot data
plt.figure(figsize=(9,8))
plt.errorbar(x = x, y = y,
             xerr = None,
             yerr = yerr,
             fmt = 'k.')
# plot model
plt.plot(x, line(m_true, x, b_true), ls='-', color='grey', label='true model')
plt.plot(x, line(m_ls, x, b_ls), ls='--', color='k', label='least squares model')
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.legend(loc='best', numpoints=1)
plt.show()
```

# Maximum Likelihood Estimation

The optimize module minimizes functions.
We want to maximize the likelihood function. 
To maximize the likelihood, you minimize the negative log likelihood:
    
    -1 * loglikelihood = Maximum Likelihood


```python
def lnlike(theta, x, y, yerr):
    m, b, lnf = theta
    model = m * x + b
    inv_sigma2 = 1.0/(yerr**2 + model**2*np.exp(2*lnf))
    return -0.5*(np.sum((y-model)**2*inv_sigma2 - np.log(inv_sigma2)))

def nll(*args):
    return -1*lnlike(*args)

result = minimize(fun=nll, 
                  x0 = [m_true, b_true, np.log(f_true)], 
                  args = (x, y, yerr))
print(result)

m_ml, b_ml, lnf_ml = result["x"]
```

```python
# plot data
plt.figure(figsize=(9,8))
plt.errorbar(x = x, y = y,
             xerr = None,
             yerr = yerr,
             fmt = 'k.')
# plot model
plt.plot(x, line(m_true, x, b_true), ls='-', color='grey', label='true model')

plt.plot(x, line(m_ls, x, b_ls), ls='--', color='k', label='least squares model')

plt.plot(x, line(m_ml, x, b_ml), ls='--', color='red', label='maximum likelihood model')

plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.legend(loc='best', numpoints=1)
plt.show()
```

# Marginalization & uncertainty estimation


```python
def lnprior(theta):
    m, b, lnf = theta
    if -5.0 < m < 0.5 and 0.0 < b < 10.0 and -10.0 < lnf < 1.0:
        return 0.0
    return -np.inf
```

```python
def lnprob(theta, x, y, yerr):
    lp = lnprior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + lnlike(theta, x, y, yerr)
```

```python
ndim, nwalkers = 3, 100
pos = [result["x"] + 1e-4*np.random.randn(ndim) for i in range(nwalkers)]
```

```python
import emcee
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob, args=(x, y, yerr))
```

```python
sampler.run_mcmc(pos, 500)
```
